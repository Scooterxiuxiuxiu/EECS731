```python
gdp2018_mean = wealth.groupby('Country Name')['2018'].mean().sort_values(ascending=False).to_frame().reset_index()
gdp2018_mean.head(50)
```

```python
wealth2018=wealth.loc[:,['Country Name','2018']]
wealth2018
```

```python
education2018=education.loc[:,['Country Name','2018']]
education2018
```

```python
data=pd.merge(education2018,wealth2018, on='Country Name')
data
```

```python
data2018=data.rename(columns={'2018_x':'SchoolEnroll','2018_y':'GDP'})
```
